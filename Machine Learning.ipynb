```python
retention_rate = df.groupby('department')['left'].mean()
```

```python
retention_rate_sorted = retention_rate.sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=retention_rate_sorted.index, y=retention_rate_sorted, color='gray')
plt.title('Retention Rate by Department')
plt.xlabel('Department')
plt.ylabel('Retention Rate')
plt.show()
```

This bar graph shows the retention rate by department. The department with the highest retention rate is HR, followed by ACCOUNTING and TECHNICAL. However, the department with the lowest retention rate is MANAGEMENT, with only 30% of employees staying with the company for more than a year.



```python
salary_retention = df.groupby('salary')['left'].mean()
salary_retention_sorted = salary_retention.sort_values(ascending=False)
plt.figure(figsize=(8,6))
sns.barplot(x=salary_retention_sorted.index, y=salary_retention_sorted, color='gray')
plt.title('Retention Rate by Salary')
plt.xlabel('Salary')
plt.ylabel('Retention Rate')
plt.show()
```

Employees low income salary are likely to leave the company. 

```python
promotion_retention_rate = df.groupby('promotion')['left'].mean()
plt.figure(figsize=(4,6))
sns.barplot(x=promotion_retention_rate.index, y=promotion_retention_rate, color='gray')
plt.title('Retention Rate by Promotion')
plt.xlabel('Promotion')
plt.ylabel('Retention Rate')
plt.show()
```

Employees with no promotion is more likely to change the company

```python

retention_rate = df.groupby('projects')['left'].mean()
plt.figure(figsize=(8,6))
sns.barplot(x=retention_rate.index, y=retention_rate, color='gray')
plt.title('Retention Rate by Number of Projects')
plt.xlabel('Number of Projects')
plt.ylabel('Retention Rate')
plt.show()
```

Employees with more projects are more likely to change the company.

#### <span style='font-family:Arial Black'>**LOGISTIC REGRESSION**</span>



```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()

df = pd.get_dummies(df, columns = (['salary']))
```

```python
y = df.left
X = df.drop(columns = ['left','department','accident'])
```

```python
Xtrain , Xtest , ytrain , ytest = train_test_split(X,y, test_size = .2)
```

```python
model.fit(Xtrain, ytrain)
```

```python
predicted = model.predict(Xtest)
predicted
```

```python
model.score(Xtest,ytest)
```
